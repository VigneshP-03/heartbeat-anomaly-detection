- Importing necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

- Loading two CSV files ('ptbdb_normal.csv' and 'ptbdb_abnormal.csv') as dataframes containing ECG data for normal and abnormal heartbeats.

```python
dataframes = [pd.read_csv('data/ptbdb_' + x + '.csv') for x in ['normal', 'abnormal']]
```

- To handle data conveniently, we rename the columns of a dataframes with consecutive integers (0 to 187)

```python
for dataframe in dataframes:
    dataframe.columns = list(range(len(dataframe.columns)))
```

- Shuffling the rows randomly, and reset the index of the combined dataframe
- Renaming last column as "Label"
- Assuming it should contain the labels for the binary classification (normal or abnormal)

```python
data = pd.concat(dataframes, axis=0).sample(frac=1.0, random_state=1).reset_index(drop=True)
data = data.rename({187: 'Label'}, axis=1)
```

- Visualising the data

```python
data
```

# Preprocessing the data
- y is assigned the 'Label' column (target variable)
- X is assigned the features by dropping the 'Label' column
- Splitting the data into training and testing sets
- 70% of the data will be used for training

```python
y = data['Label'].copy()
X = data.drop('Label', axis=1).copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=1)
```

# Training the model
- Defining input layer of the neural network with the shape equal to the number of features in X_train
- Adding an extra dimension to the input data, for processing time series data in neural networks
- Adding a Gated Recurrent Unit (GRU) layer with 256 units to the model, suitable for processing sequential data
- Adding a Dense layer with one output unit and a sigmoid activation function, making it suitable for binary classification

```python
inputs = tf.keras.Input(shape=(X_train.shape[1],))

expand = tf.expand_dims(inputs, axis=2)
gru = tf.keras.layers.GRU(256, return_sequences=True)(expand)
flatten = tf.keras.layers.Flatten()(gru)

outputs = tf.keras.layers.Dense(1, activation='sigmoid')(flatten)
```

- Creating a Keras model with specified input and output layers

```python
model = tf.keras.Model(inputs=inputs, outputs=outputs)

print(model.summary())
```

- Compiling model using Adam optimizer, and we use 20% of the data for validation.
- Training is done for 20 epochs with early stopping to prevent overfitting.

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ]
)

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=20,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
    ]
)
```

- Results evaluation

```python
results = model.evaluate(X_test, y_test, verbose=0)
```

```python
print("Test Accuracy: {:.2f}%".format(results[1] * 100))
print("Test AUC: {:.4f}".format(results[2]))
```
